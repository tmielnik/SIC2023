# Hello-World: Redes Neurais

```python
# Imports
import tensorflow as tf
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
%matplotlib ipympl
```

```python
# Exibe a versão instalada do tensorflow
print(tf.__version__)
```

```python
# Cria um conjunto de dados simples com as entradas e suas respectivas saídas
inputs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)  # X
labels = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float) # Y

# Percebe-se que a saída de sua respectiva entrada segue o padrão: 2X - 1 (relação linear)
print(inputs)
print(labels)
```

```python
# Cria um modelo simples de rede neural com 1 neurônio
model = keras.Sequential(name='Hello_World_ANN')
model.add(layers.Dense(units=1, activation='linear', input_shape=(1,)))

# Compila o modelo definindo a função de perda e o algoritmo otmizador dos pesos
model.compile(loss='mse', optimizer='sgd')
```

```python
# Exibe algumas informações do modelo
print(model.summary())
```

```python
%%time

# Treina o modelo por 500 épocas (treina sobre todo o conjunto de dados 500 vezes)
history = model.fit(inputs, labels, epochs=500, verbose=1)
```

```python
# Mostra o gráfico do erro da rede neural
plt.figure(figsize=(8,4))
plt.plot(history.history["loss"])
plt.legend(["Loss training"])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid()

plt.show()
```

```python
# Faz a predição de um número com base nas regras/padrões encontradas durante o treinamento
print(model.predict([10]))
```

```python
# Finaliza as figuras
plt.close('all')
```

```python
# Mostra o peso e bias do neurônio
print(model.weights)
```
